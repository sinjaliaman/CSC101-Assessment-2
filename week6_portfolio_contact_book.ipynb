```python
# CSC101 Assessment 2: Part A

# Create contact dictionary
contacts = {
    "luffy" : "0400 123 456",
    "Ace" : "0500 654 321",
    "sabo" : "0600 987 654",
}

# Add a new entry to the dictionary
contacts["zoro"] = "0700 222 333"

# Update the value of an existing entry
contacts["Ace"] = "1222 040 303"

# Print the key–value pairs
for key, value in contacts.items():
    print(f"{key} : {value}")
```
