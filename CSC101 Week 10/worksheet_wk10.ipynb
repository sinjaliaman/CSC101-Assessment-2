1.1 Problem 1: Tip Calculator
1.1.1 The Scenario
You work at a Perth restaurant. The manager wants a simple tool where staff can enter:

The bill amount (e.g., $45.80)
The tip percentage (e.g., 15%)
And get back:

The tip amount
The total amount to charge
1.1.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

[Answer: The core expectation of the program is to take two numbers, which are the bill and tip, and evaluate two results. The program
 should ask the user to enter the bill amount and the program stores this as a number. Next, the user enters the tip amount and
 the program converts the '%'  into a decimal. Finally, the tip amount and total bill will be calculated seperately and displayed.]

STEP 2: Identify Inputs and Outputs

INPUTS:

[What does the user provide? Answer: The user provides the bill and the tip '%'.]
[What type of data? Numbers? Text? Answer: The data are numbers.]
OUTPUTS:

[What does the program show back? Answer: The tip amount and the total bill to be charged.]
[What format? Answer: The final output is displayed in dollar amount(currency format).]
STEP 3: Work It By Hand

Show me you can do this with a calculator. Use this example:-

Bill amount: \$45.80
Tip percentage: 15\%
Write the steps:

[1st Calculation: Convert 15\% into decimal; 15/100 = 0.15, 
 2nd Calculation: Calculate tip amount; \$45.80 x 0.15 = \$6.87,
 Final Calculation: Calculate bill and tip for final result; \$45.80 + \$6.87 = \$52.67,
 What I would tell the staff member. Answer: The \$45.80 bill has a 15\% tip of \$6.87. Total amount to be charged is \$52.67]

1.2 Problem 2: Late Fee Calculator
1.2.1 The Scenario
You work for a library in Perth. Books are due back in 14 days. For every day late, there’s a /$0.50 fine (maximum /$10.00).

The librarian wants a program where they enter:

How many days the book is overdue
And get back:

The fine amount (or “Maximum fine reached” if it’s over \$10)
1.2.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

[Answer: This program will assist the librarian to calculated the amount a borrower owes if a book is returned late. The user will
         enter how many days overdue the book is, and the program will calculate the fine based on \$0.50/day, without exceeding the 
         maximum amount of \$10.00. If the fine reaches the maximum, the program should show that the maximum fine has been reached.   ]

STEP 2: Identify Inputs and Outputs

INPUTS:

[What information does the librarian provide? Answer: The librarian should provide the no. of days overdue for example an integer like 0, 3, 10, etc.]
OUTPUTS:

[What does the program calculate and show? Answer: The total fine amount will be shown following the \$0.50/day rule.]
[Any special cases? Answer: If the fine ever reaches the maximum cap of \$10, the program will display "Maximum amount reached: \$10".]
